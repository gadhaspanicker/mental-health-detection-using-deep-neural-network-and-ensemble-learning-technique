Label encoding

```python
data = data.drop('Timestamp', axis=1)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
cols_to_encode = ['self_employed', 'family_history', 'treatment',
                  'Growing_Stress', 'Changes_Habits', 'Mental_Health_History',
                  'Mood_Swings', 'Coping_Struggles', 'Work_Interest',
                  'Social_Weakness', 'mental_health_interview', 'care_options']

le = LabelEncoder()
for col in cols_to_encode:
    data[col] = le.fit_transform(data[col].astype(str))
```

```python
data['Days_Indoors'] = data['Days_Indoors'].map({
    '1-14 days': 7,
    '15-30 days': 22,
    'More than 30 days': 45
})
```

```python
data['Gender'] = le.fit_transform(data['Gender'].astype(str))
data['Country'] = le.fit_transform(data['Country'].astype(str))
data['Occupation'] = le.fit_transform(data['Occupation'].astype(str))
```

```python
data.head()
```

```python
# Check for missing values in 'Days_Indoors'
print(data['Days_Indoors'].isnull().sum())

# Fill missing values with the median of the column
data['Days_Indoors'].fillna(data['Days_Indoors'].median(), inplace=True)
```

feature selection

```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import matplotlib.pyplot as plt

# Split features and target
X = data.drop('treatment', axis=1)
y = data['treatment']

# Fit model
model = RandomForestClassifier()
model.fit(X, y)

# Get feature importances
importances = model.feature_importances_
features = X.columns

# Plot feature importances
plt.figure(figsize=(10, 6))
plt.barh(features, importances)
plt.xlabel("Feature Importance")
plt.title("Feature Selection using Random Forest")
plt.show()

# Print top features
importance_df = pd.DataFrame({'Feature': features, 'Importance': importances})
print(importance_df.sort_values(by='Importance', ascending=False))
```

```python
# Manually list features with importance > 0.01
selected_features = [
    'Country',
    'family_history',
    'care_options',
    'mental_health_interview',
    'self_employed',
    'Gender',
    'Occupation',
    'Social_Weakness',
    'Mood_Swings',
    'Work_Interest',
    'Changes_Habits'
]

# Keep only selected features (plus target)
x = data[selected_features]
y = data['treatment']  # Assuming 'treatment' is your target

# Optional: Combine selected features and target
df = pd.concat([x, y], axis=1)
```

ROC and AOC

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Predict probabilities
#y_prob_ann = model_ann.predict(x_test).ravel()  # ANN gives ravelled output
y_prob_xgb = xgb_model.predict_proba(x_test)[:, 1]
y_prob_rf = rf_model.predict_proba(x_test)[:, 1]
y_prob_knn = model.predict_proba(x_test)[:, 1]
y_prob_ada = ada_model.predict_proba(x_test)[:, 1]

# Get ROC values
#fpr_ann, tpr_ann, _ = roc_curve(y_test, y_prob_ann)
fpr_xgb, tpr_xgb, _ = roc_curve(y_test, y_prob_xgb)
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_prob_rf)
fpr_knn, tpr_knn, _ = roc_curve(y_test, y_prob_knn)
fpr_ada, tpr_ada, _ = roc_curve(y_test, y_prob_ada)

# Get AUC scores
#auc_ann = auc(fpr_ann, tpr_ann)
auc_xgb = auc(fpr_xgb, tpr_xgb)
auc_rf = auc(fpr_rf, tpr_rf)
auc_knn = auc(fpr_knn, tpr_knn)
auc_ada = auc(fpr_ada, tpr_ada)

# Plot all ROC curves
plt.figure(figsize=(10, 7))
#plt.plot(fpr_ann, tpr_ann, label='ANN (AUC = {:.2f})'.format(auc_ann))
plt.plot(fpr_xgb, tpr_xgb, label='XGBoost (AUC = {:.2f})'.format(auc_xgb))
plt.plot(fpr_rf, tpr_rf, label='Random Forest (AUC = {:.2f})'.format(auc_rf))
plt.plot(fpr_knn, tpr_knn, label='KNN (AUC = {:.2f})'.format(auc_knn))
plt.plot(fpr_ada, tpr_ada, label='AdaBoost (AUC = {:.2f})'.format(auc_ada))

plt.plot([0, 1], [0, 1], 'k--', label='Chance')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Comparison for All Models')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```
